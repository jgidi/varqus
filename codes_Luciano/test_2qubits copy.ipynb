```python
import numpy as np
import matplotlib.pyplot as plt
from qiskit import Aer
from main import *
```

```python
simulator = Aer.get_backend('aer_simulator')
shots = 2**13
steps = 1
params = [0, 0]
```

```python
R = [ [ 'ry', [0] ], 
         [ 'cx', [0,1] ], 
         [ 'rz', [1] ] ]
Uk = [ [ 'cy', [0] ],   
        None , 
        [ 'cz', [1] ] ]
Uq = [ [ 'cy', [0] ], 
        None , 
        [ 'cz', [1] ], 
        [ 'cyy', [0,1] ]  ]
```

```python
circ = circuits( R, Uk, Uq, 2 )
```

```python
circ[2].draw(output='mpl')
```

```python
circ_par = [ qc.assign_parameters(params) for qc in circ  ]
job = simulator.run( circ_par , shots=shots )
counts = job.result().get_counts()[2]
( counts['0'] - counts['1'] ) / shots
```
